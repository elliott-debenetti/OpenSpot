# Plot a single model

```python
# model_name = 'RCNN_FPN_800_square'
train_log = pd.read_csv('/out_dir/RCNN_v2/train_log.csv')
va = train_log.valid_accuracy.tolist()
ta = train_log.train_accuracy.tolist()


plt.plot(va, label='Validation')
plt.plot(ta, label='Training')

plt.legend()

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.show()
```
